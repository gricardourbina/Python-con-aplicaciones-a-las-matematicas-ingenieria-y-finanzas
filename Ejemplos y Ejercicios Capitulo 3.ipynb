Ejemplo 3.8

```python
#Ecuación cuadrática
from math import *
a=float(input("Ingrese coeficiente a: "))
b=float(input("Ingrese coeficiente b: "))
c=float(input("Ingrese coeficiente c: "))
d=pow(b,2)-4*a*c
if (d==0):
    print("Raíces iguales")
    x1=-b/(2*a)
    x2=x1
    print(x1)
if (d>0):
    print("Raíces reales y diferentes")
    x1=(-b+sqrt(d))/(2*a)
    x2=(-b-sqrt(d))/(2*a)
    print("x1=", x1) 
    print("x2=",x2)
if (d<0):
    print("Raíces complejas")
    x1_real=-b/(2*a)
    x1_imaginaria=sqrt(-d)/(2*a)
    x2_real=x1_real
    x2_imaginaria=-sqrt(-d)/(2*a)
    print(x1_real, x1_imaginaria)
    print(x2_real, x2_imaginaria)
```

```python
K=0.0025
Vt=0.25
Vgs=0.5
Vds=float(input("Ingrese Vds (0.5,1): "))
if (Vds!=0.5 and Vds!=1):
    print("No es un valor permitido para Vds")
else:
    if Vds>=(Vgs-Vt):
        Id=K*(pow((Vgs-Vt),2))
        print("Id=%0.9f" %Id)
    else:
        Id=K*((2*(Vgs-Vt)*Vds)-pow(Vds,2))
        print("Id=%0.9f" %Id)
```
